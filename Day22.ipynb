```python
import pandas as pd

data = {
    'Neighborhood': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Topping': ['Pepperoni', 'Cheese', 'Cheese', 'Mushroom', 'Pepperoni', 'Cheese'],
    'Slices': [10, 5, 15, 5, 20, 10]
}
df = pd.DataFrame(data)
df
```

```python

total_slices = df.groupby('Neighborhood')['Slices'].sum().reset_index()
total_slices
```

```python
total_slices_by_topping = df.groupby(['Neighborhood', 'Topping'])['Slices'].sum().reset_index()
total_slices_by_topping
```
